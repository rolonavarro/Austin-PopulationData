```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# URL: https://data.austintexas.gov/Health-and-Community-Services/Strategic-Measure_Number-of-persons-experiencing-h/infv-inkz/data

path = '../../Appendix_A_Resources/Strategic_Measure_Number_of_persons_experiencing_homelessness_-_Point_in_Time_Count.csv'

homeless_df = pd.DataFrame(pd.read_csv(path))

homeless_df = homeless_df.sort_values('PIT Count Year')

homeless_df
```

```python
year = []

# A loop to get date down to just year
for index, row in homeless_df.iterrows():
    year.append(row['PIT Count Year'][6:10])
    
sheltered = homeless_df['Sheltered']
unsheltered = homeless_df['Unsheltered']
total = homeless_df['Total Unduplicated Individuals']
```

```python
# Plot homeless count over time
plt.figure(figsize=(12,8))
plt.plot(year, total)
plt.title('Austin Homeless Count (2010-2018)')
plt.xlabel('Year')
plt.ylabel('Number of Persons Experiencing Homelessness')
plt.ylim(0,3000)

# Save to Output_Data
plt.savefig('../../Appendix_B_Output_Data/homeless_austin.png', dpi=100)
```

```python
# Plot sheltered and unsheltered homeless populations over time
plt.figure(figsize=(12,8))
plt.plot(year, unsheltered, label='Unsheltered')
plt.plot(year, sheltered, label='Sheltered')
plt.title('Austin Homeless Count (2010-2018)')
plt.xlabel('Year')
plt.ylabel('Number of Persons Experiencing Homelessness')
plt.ylim(0,2000)
plt.legend()

# Save to Output_Data
plt.savefig('../../Appendix_B_Output_Data/homeless_sheltered.png', dpi=100)
```
