```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read CSV into DF
path = '../../Appendix_A_Resources/Quick_pop_data/Equity___Livability_-_Median_Household_Income_by_race_ethnicity.csv'

equity_df = pd.DataFrame(pd.read_csv(path))

equity_df.head()
```

```python
# Set desired column as variable (saves space)
column = 'Key Performance Indicator'

# Get years and income values for each category
year = aa_income = equity_df.loc[equity_df[column] == 'Median Household Income ($) - African American']['Chart date']
aa_income = equity_df.loc[equity_df[column] == 'Median Household Income ($) - African American']['Value']
asian_income = equity_df.loc[equity_df[column] == 'Median Household Income ($) - Asian']['Value']
lat_income = equity_df.loc[equity_df[column] == 'Median Household Income ($) - Hispanic/Latino']['Value']
white_income = equity_df.loc[equity_df[column] == 'Median Household Income ($) - White']['Value']
```

```python
# Plot each category as its own line
plt.figure(figsize=(12,8))
plt.plot(year, aa_income, label='African American')
plt.plot(year, asian_income, label='Asian', c='purple')
plt.plot(year, lat_income, label='Latinx', c='g')
plt.plot(year, white_income, label='White', c='r')
plt.title('Median Household Income by Race Ethnicity (2011-2016)')
plt.xlabel('year')
plt.ylabel('Median Household Income ($)')
plt.ylim(30000,85000)
plt.legend()

# Save to Output_Data
plt.savefig('../../Appendix_B_Output_Data/income_race.png', dpi=100)
```
